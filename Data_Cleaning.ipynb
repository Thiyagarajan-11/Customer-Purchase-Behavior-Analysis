```python
import pandas as pd

df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.rename(columns={"Promo Code Used":"Promo"},inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))    
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df.rename(columns={"purchase_amount_(usd)":"purchase_amount"},inplace=True)
```

```python
df.columns
```

```python
l=['Young Adult','Adult','Middle-age','Senior']
df['age_group']=pd.qcut( df['age'],q=4,labels=l)
```

```python
df[['age_group','age']].tail(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
fre_map={
    'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':180, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}
df['purchase_frequency_day']=df['frequency_of_purchases'].map(fre_map)
```

```python
df[['frequency_of_purchases','purchase_frequency_day']].head()
```

```python
df.head()
```

```python
df=df.drop('promo',axis=1)
```

```python
df.to_csv("Cleaned_data.csv",index=False)
```
